<a name="002"></a><h2>Створення множества з текстової строки яка записана в клітинку Excel</h2>
<a href="#1">На початок</a>

---
*(Методи та оператори множеств стор.609 Марк Саммерфилд программирование на Python 3)*
```Python
import pandas as pd
import re
from openpyxl import load_workbook


wb = load_workbook('амб_вилучені.xlsx')
ws = wb.active

def convert_set_re(sh, col, start, finish):
    """
    Повертає множество створено зі значень клітинок файлу excel
    sh - активна сторінка файлу excel,
    col - номер колонки активної сторінки файлу excel,
    start - номер рядка початка колонки активної сторінки файлу excel,
    finish - номер рядка кінця колонки активної сторінки файлу excel
    """
    r = set()
    for i in range(start, finish + 1):
        c = sh.cell(row=i, column=col).value
        # текс в клітинці подібен "Y38002, B34029, 97048-01, 97047-1"
        rc1 = re.findall(r'[A-ZА-Я]\d{5}', c)
        rc2 = re.findall(r'\d{5}-\d\d?', c)
        rc = rc1 + rc2
        r = r.union(rc)  
    return r


# створення множества за допомогою функції convert_set_re()
# з різних колонок файла excel
set_22 = convert_set_re(sh=ws, col=3, start=2, finish=12)
set_23 = convert_set_re(sh=ws, col=4, start=2, finish=12)
# запис множества у текстовий файл
s1 = ','.join(list(set_22))
with open("set_22.txt", "w") as file:
    file.write(s1)
# запис множества у текстовий файл
s2 = ','.join(list(set_23))
with open("set_23.txt", "w") as file:
    file.write(s2)
    
# Далі можно завантажити ці файли у power query, скасувати зведення
# стовпців та зробити об'єднання файлів
```
<h4>Альтернативний варіант функції</h4>

Правильно спрацює лише коле елементи будуть відділятись ", "  _(кома з пробілом)_
    
```Python
def convert_set(sh, col, start, finish):
    """
    Повертає множество створено зі значень клітинок файлу excel
    sh - активна сторінка файлу excel,
    col - номер колонки активної сторінки файлу excel,
    start - номер рядка початка колонки активної сторінки файлу excel,
    finish - номер рядка кінця колонки активної сторінки файлу excel,
    """
    lst = list()
    for i in range(start, finish + 1):
        c = sh.cell(row=i, column=col).value
        for item in  c.split(", "):
            lst.append(item)
    
    return set(lst)    
```    

<a name="003"></a><h2>Видаленя дублікатів діагнозів,  записанних в текстову строку.</h2>
<a href="#1">На початок</a>

---
<h4>Функції для считування текстового файлу</h4>

- readline(): считывает одну строку из файла
- read(): считывает все содержимое файла в одну строку
- readlines(): считывает все строки файла в список


<h4>Варіант 1.</h4> За допомогою set()

```Python
# змінна для ідентифікації файлу
share_nm_fale = 1
# шлях до файлу який треба прочитати
file = f"./data/K{share_nm_fale}.txt"
# шлях до файлу який потрібно записати (результат)
file_out = f"KK{share_nm_fale}.txt"

# считування файлу однією строкою в змінну s та видаленя
# /n чи пробілу (якщо є)
f = open(file, encoding='utf-8')
s = f.read().rstrip()
f.close()

# перетворення строки спочатку в список далі в множество (для видалення дублікатів 
# елементів списку)
sett = set(s.split(", "))

# перетворення множества в список далі в строку
result = ', '.join(sorted(list(sett)))
# запис строки result в файл
with open(file_out, "w", encoding='utf-8') as file:
    file.write(result)
```
<h4>Варіант 2.</h4> За допомогою перебору елементів list()

```Python
share_nm_fale = 2
# шлях до файлу який треба прочитати
file = f"./data/К{share_nm_fale}.txt"
# шлях до файлу який потрібно записати (результат)
file_out = f"KK{share_nm_fale}.txt"

# считування файлу однією строкою в змінну s та видаленя
# /n чи пробілу (якщо є)
f = open(file, encoding='utf-8')
s = f.read().rstrip()
f.close()
# перетворення строки в відсортованний список
lst = sorted(s.split(", "))
# заповнення нового списку унікальними елементами
li = []
for i in lst:
    if i not in li:
        li.append(i)


# перетворення списку li в строку та запис у файл file_out      
result = ', '.join(li)
with open(file_out, "w", encoding='utf-8') as fіlе:
    fіlе.write(result)
```

<h4>Приклад цикла для обробки файлів K1.txt, K2.txt, K3.txt, K4.txt</h4>

```Python
for j in range(1, 5):
    share_nm_fale = j
    file = f"./data/K{share_nm_fale}.txt"
    file_out = f"KK{share_nm_fale}.txt"

    f = open(file)
    s = f.read().rstrip()
    f.close()
    lst = sorted(s.split(", "))
    li = []
    for i in lst:
        if i not in li:
            li.append(i)

    result = ','.join(li)
    with open(file_out, "w") as fіlе:
        fіlе.write(result)
```      

<a name="004"></a><h2>Створеня словаря python з текстового файлу.</h2>
<a href="#1">На початок</a>

**Є файл з таким змістом:**<br/>
transfer,Переведений з іншого ЗОЗ<br/>
emergency,Бригадою екстреної (швидкої) медичної допомоги<br/>
born_in_LE,Народжений у закладі<br/>
third_party,Доставлений третіми особами<br/>
self_сonvers,Самозвернення<br/>
blank_referral,За направленням паперовим<br/>
transfer_in_LE,Переведено з іншого відділення<br/>
system_referral,За направленням електронним<br/>

```Python
import pandas

dct = dict()
with open("./data/admission_source.csv", encoding='utf-8') as f:
    lst = [i.rstrip() for i in list(f)]
    for j in lst:
        dct[j.split(",")[0]] = j.split(",")[1]
```

<a name="005"></a><h2>Setdefault()</h2>
<a href="#1">На початок</a>

Метод setdefault() возвращает значение элемента с указанным ключом. Если ключ не существует, вставляет ключ с указанным значением.

```Python
lst = ['a','b','b','c','a','a','a']
dc = dict()
for key in lst:
    val = dc.setdefault(key, 0)
    dc[key] = val + 1

print(dc)

{'a': 4, 'b': 2, 'c': 1}
```
